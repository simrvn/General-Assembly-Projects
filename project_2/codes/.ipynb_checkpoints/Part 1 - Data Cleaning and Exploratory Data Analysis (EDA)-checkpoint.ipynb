# Project 2: Ames Housing Data and Kaggle Challenge

## Introduction

The aim of this project is to determine the best regression model that is able to predict prices of houses in Ames accurately.

## Executive Summary





### Methodology 
The project is split into 4 sections which are further described below.

1. Data Cleaning and Feature Engineering

    - This involved examining the data and looking for missing values for both the Train and Test Datasets.
    - Followed by feature classification, where we categorise the features based on their datatypes and analyse them separately.
    - Exploratory Data Analysis was conducted on the categorised features as well to look for features that were highly skewed, as well as outliers and/or peculiar data.
    - Outliers were identified and removed if deemed necessary
    - Muliticollinerity was identified amongst features and removed to prevent any form of penalisations to our models later.
    

2. Data Preprocessing

    - One hot encoding was done to our nominal features so that we can model them later. 
    - Ordinal and Categorical data were scaled thereafter
    - Scaled continuous and ordinal features
    

3. Modelling and Hyperparameter tuning

    - Regression Models were created to identify the best model to predict prices. Lasso Regression Model held the lowest RMSE and hence is the best model
    
    
4. Best Model and Kaggle Submission

    - Model was implemented on Test Dataset and submitted to Kaggle. Kaggle score was turned out to be higher.

Functions for plots.

```python
#Boxplot for comparing single features.

def box_plot(x_col, y_col, palette, title):
    fig, ax = plt.subplots(figsize = (8,6))
    sns.boxplot(x = x_col, y = y_col, data = train, ax=ax, palette = palette)
    plt.xlabel(x_col, fontsize = 16)
    plt.ylabel(y_col, fontsize = 16)
    plt.title(title, fontsize = 18)
    plt.show()
```

```python
#scatterplot
def subplot_scatter(df, list_of_columns, title=None):
    nrows = int(np.ceil(len(list_of_columns)/2))
    fig, ax = plt.subplots(nrows=nrows, ncols=2, figsize = (15, nrows*5), sharex=False, sharey=True)
    ax = ax.ravel() 
    for i, column in enumerate(list_of_columns): 
        sns.regplot(x = df[column], y = df.saleprice, ax = ax[i], line_kws={'color': 'red'}) 
        ax[i].set_title(column,fontsize=18)
        ax[i].tick_params(labelsize=12)
        ax[i].set_xlabel(column, fontsize=12)
        ax[i].set_ylabel('saleprice', fontsize=12)
    plt.title(title, fontsize = 18)
    fig.tight_layout()


#boxplot
def subplot_boxplot(df, list_of_columns, title=None):
    nrows = int(np.ceil(len(list_of_columns)/4))
    fig, ax = plt.subplots(nrows=nrows, ncols = len(list_of_columns), figsize =(15, nrows*5), 
                           sharex=False, sharey=True)
    ax = ax.ravel() 
    for i, column in enumerate(list_of_columns): 
        sns.boxplot(x = df[column], y = df.saleprice, ax = ax[i]) 
        ax[i].set_title(column,fontsize=18)
        ax[i].tick_params(labelsize=12)
        ax[i].set_xlabel(column, fontsize=12)
    plt.title(title, fontsize = 18)
    fig.tight_layout()


#histogram
def subplot_hist(df, list_of_columns, title=None):
    nrows = int(np.ceil(len(list_of_columns)/4)) 
    fig, ax = plt.subplots(nrows=nrows, ncols=4,figsize =(15, nrows*5)) 
    ax = ax.ravel() 
    for i, column in enumerate(list_of_columns): 
        ax[i].hist(df[column],bins=15)
        ax[i].set_title(f'{column} distribution',fontsize=18)
        ax[i].tick_params(labelsize=12)
        ax[i].set_xlabel(column, fontsize=12)
    plt.title(title, fontsize = 18)
    fig.tight_layout()
    

#histogram and boxplot
def subplot_box_hist(df, list_of_columns, title=None):
    nrows = len(list_of_columns)
    fig, ax = plt.subplots(nrows=nrows, ncols=2,figsize =(15, nrows*5)) 
    ax = ax.ravel() 
    for i, column in enumerate(list_of_columns):
        sns.boxplot(ax=ax[i*2], x=column, y='saleprice', data=df)
        
        if df[column].dtype == object:
            sns.histplot(ax=ax[(i*2 + 1)], x=column, data=df, stat='probability', legend=False)
        else: 
            sns.histplot(ax=ax[(i*2 + 1)], x=column, data=df, legend=False, stat='probability', 
                         discrete=True)
            
        ax[i*2].set_xlabel(column, fontsize=12)
        ax[(i*2 + 1)].set_xlabel(column, fontsize=12)
        ax[(i*2 + 1)].axhline(0.90, ls='--')
        
    fig.tight_layout()
    
def heatmap(df, title=None, annot=False):
    # plot heatmap
    mask = np.zeros_like(df.corr(), dtype=np.bool)
    mask[np.triu_indices_from(mask)]= True

    fig, ax = plt.subplots(figsize=(15, 15))
    sns.heatmap(df.corr().abs(),  mask = mask, square=True,vmax = 1, 
            linewidth=0.1, cbar_kws = {'shrink':0.5}, cmap='rocket_r', annot=annot)
    ax.tick_params(labelsize=12)
    plt.tight_layout()
    plt.title(title, fontsize = 16);
```

The features `overall_qual`, `overall_cond`, `bsmt_full_bath`, `bsmt_half_bath`, `full_bath`, `half_bath`, `bedroom_abvgr`, `kitchen_abvgr`, `totrms_abvgrd`, `fireplaces`, `garage_cars` hold numerical values, so we do not need to map them.

bsmt_qual, bsmt_cond, heating_qc, fireplace_qu, garage_qual, garage_cond appear to have a similar order. We will map them together. Let's store them in a list first.

We will also map each feature according to the standards in the data dictionary where `overall_qual` and `overall_cond` have been ranked from 1 to 10 with 1 representing the lowest rank and 10 the highest. 'NA' will be represented by 0.

```python
#function to map values for ordinal features
temp = {'NA': 0, 'Po':1, 'Fa': 2, 'TA': 3, 'Gd': 4, 'Ex': 5}
def ordinal_map(df, feature_list, dictionary):
    for feature in feature_list:
        df[feature] = df[feature].replace(dictionary)
```

```python
#store similar ordered columns in same list
feat_1 = ['bsmt_qual', 'bsmt_cond', 'heating_qc', 'fireplace_qu', 'garage_qual', 'garage_cond',
         'exter_qual', 'exter_cond', 'kitchen_qual', 'pool_qc']
```

```python
#map the values in columns in feat_1 list in train dataset
ordinal_map(train, feat_1, temp)
```

For the remaining columns we will have to do them individually.

```python
#lot_shape
lot_dict = {'IR3':0, 'IR2':1, 'IR1':2, 'Reg':3}
train['lot_shape'] = train['lot_shape'].replace(lot_dict)         

#utilities
util_dict = {'ELO':0, 'NoSeWa':1, 'NoSewr':2, 'AllPub':3}
train['utilities'] = train['utilities'].replace(util_dict)               

#land_slope
slope_dict = {'Gtl':0, 'Mod':1, 'Sev':2 }
train['land_slope'] = train['land_slope'].replace(slope_dict)               

#bsmt_exposure
expo_dict = {'NA':0,'No':1, 'Mn':2, 'Av': 3, 'Gd':4}
train['bsmt_exposure'] = train['bsmt_exposure'].replace(expo_dict)  

#bsmtfin_type_1
type1_dict = {'NA':0, 'Unf':1, 'LwQ':2, 'Rec':3, 'BLQ':4, 'ALQ': 5, 'GLQ':6}
train['bsmtfin_type_1'] = train['bsmtfin_type_1'].replace(type1_dict)

#bsmtfin_type_2
type2_dict = {'NA':0, 'Unf':1, 'LwQ':2, 'Rec':3, 'BLQ':4, 'ALQ': 5, 'GLQ':6}
train['bsmtfin_type_2'] = train['bsmtfin_type_2'].replace(type2_dict) 

#electrical
elec_dict= {'Mix':0, 'FuseP': 1, 'FuseF': 2, 'FuseA': 3,'SBrkr':4} 
train['electrical'] = train['electrical'].replace(elec_dict)   

#paved_drive
paved_dict = {'N':0, 'P':1, 'Y':2 }
train['paved_drive'] = train['paved_drive'].replace(paved_dict) 

#garage_finish
gfinish_dict = {'NA':0, 'Unf':1, 'RFn':2,'Fin':3}
train['garage_finish'] = train['garage_finish'].replace(gfinish_dict) 

#functional
func_dict = {'Sal': 0, 'Sev':1, 'Maj2':2, 'Maj1':3, 'Mod':4, 'Min2':5, 'Min1': 6, 'Typ':7}
train['functional'] = train['functional'].replace(func_dict)

#fence
fence_dict = {'NA':0, 'MnWw':1, 'GdWo':2, 'MnPrv':3, 'GdPrv':4}
train['fence'] = train['fence'].replace(fence_dict)
```

### Multicollinearity and Interaction Terms

There are many features that belong to similar groups and have high collinearity amongst them. We will analyse them further to decide the features we want to keep or drop to reduce collinearity. First, we will look at a heatmap to understand the collinearity amongst features.

```python
#sort columns 
df = train[train.columns.sort_values()]
title = 'Correlation Heatmap of all Features'
heatmap(df, title)
```

The darker squares are the points of interest. These represent features that are highly correlated with each other. Additionally, some of these also belong to similar groups, such as basement and porch, for example, and we will look at these more closely.

**Property Age**

We can deduce the age of a property at the time it was sold through two features `yr_sold` and `year_built` and the same for garage as well with `garage_yr_blt` and `yr_sold` and `year_remod_add` and `yr_sold` We can create new features known as `age` describing age of property, `garage_age` describing age of garage and `age_remod_add` describing age of remod/add.

```python
train['property_age'] = train['yr_sold'] - train['year_built']
train['age_remod_add'] = train['yr_sold'] - train['year_remod_add']
```

Since not every property has a garage, we will only calculate age for properties with a garage.

```python
def find_age(row):
    if row['garage_yr_blt'] == 0:
        return 0
    else:
        return row['yr_sold'] - row['garage_yr_blt']
```

```python
train['garage_age'] = train.apply(find_age, axis=1)
```

```python
continuous_drop.extend(['year_built', 'garage_yr_blt', 'year_remod_add'])
nominal_drop.extend(['yr_sold'])
continuous_feat.extend(['property_age', 'age_remod_add', 'garage_age'])
```

```python
train.drop(columns=['yr_sold', 'year_built', 'garage_yr_blt', 'year_remod_add'], inplace=True)
```
